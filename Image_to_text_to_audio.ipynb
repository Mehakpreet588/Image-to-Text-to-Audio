```python
import pytesseract
from PIL import Image
import gtts
```

```python
img = Image.open('/content/IMAGE.jpg')
print(img)
```

```python
result = pytesseract.image_to_string(img)
print(result)
tts = gtts.gTTS(result)
tts.save("Audio.mp3")
```
